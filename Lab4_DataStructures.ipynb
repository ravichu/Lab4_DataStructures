```python
# === Exercise 1 ===
from collections import defaultdict, Counter

# Step 1: Initialize inputs
SEED_NUM = 7
STUDENT_MAJOR = "BSME"
FAVORITE_ARTIST = "TJ MONTERDE"
LOCAL_HAZARD = "EARTHQUAKE"
CONTROL_NUM = max(3, SEED_NUM)

# Step 2: Base stream
stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]

# Step 3: Data Expansion (add CONTROL_NUM + 3 = 10 entries)
extra_data = [
    ("TUP Manila", "BSME"),
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", "BSME"),
    ("TUP Visayas", "BSME"),
    ("TUP Manila", "BSME"),
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", "BSME"),
    ("TUP Manila", "BSECE")
]

stream.extend(extra_data)

# Step 4: Aggregation using defaultdict
campus_data = defaultdict(list)

for campus, program in stream:
    campus_data[campus].append(program)

# Step 5: Frequency analysis using Counter
manila_counter = Counter(campus_data["TUP Manila"])

top_program, frequency = manila_counter.most_common(1)[0]

# Step 6: Outputs
print("CONTROL_NUM:", CONTROL_NUM)
print("STUDENT_MAJOR:", STUDENT_MAJOR)
print("Expanded Stream Length:", len(stream))
print("Manila Applications:", len(campus_data["TUP Manila"]))
print("Top Program Manila:", top_program)
print("Frequency:", frequency)
```

```python
# === Exercise 2 ===

# Step 1: Festival set
festival = {
    "Ben&Ben", "SB19", "Bini", "Eraserheads",
    FAVORITE_ARTIST, "Zild", f"Indie Artist {CONTROL_NUM}"
}

# Step 2: Users
user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"}

# Step 3: Add CONTROL_NUM = 7 artists
festival.update({
    "Arthur Nery", "Moira", "December Avenue",
    "Unique", "Cup of Joe", "Adie", "KZ Tandingan"
})

# Step 4: Jaccard formula
intersection_a = user_a.intersection(festival)
union_a = user_a.union(festival)

jaccard_a = (len(intersection_a) / len(union_a)) * 100

intersection_b = user_b.intersection(festival)
union_b = user_b.union(festival)

jaccard_b = (len(intersection_b) / len(union_b)) * 100

# Step 5: Dealbreakers
dealbreakers = user_a.difference(festival)

# Step 6: Output
print("Festival Size:", len(festival))
print("Intersection A:", len(intersection_a))
print("Union A:", len(union_a))
print("Jaccard A:", jaccard_a)
print("Jaccard B:", jaccard_b)
print("Dealbreakers:", dealbreakers)
```

```python
# === Exercise 3 ===
from collections import deque, Counter

# Step 1: Initialize deque
dq = deque(maxlen=5)

# Step 2: Base burst
burst = [
    (CONTROL_NUM, "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, LOCAL_HAZARD),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]

# Step 3: Add CONTROL_NUM + 2 = 9 alerts
extra_alerts = [
    (14, "WEATHER"),
    (15, "TRAFFIC"),
    (16, "WEATHER"),
    (17, LOCAL_HAZARD),
    (18, "WEATHER"),
    (19, "TRAFFIC"),
    (20, "WEATHER"),
    (21, LOCAL_HAZARD),
    (22, "WEATHER")
]

burst.extend(extra_alerts)

# Step 4: Sliding window filter
weather_at_3_appended = None
hazard_appended = None

for timestamp, category in burst:
    categories = [c for _, c in dq]
    count = Counter(categories)

    if count[category] >= 2:
        if timestamp == CONTROL_NUM + 3:
            weather_at_3_appended = "N"
        if category == LOCAL_HAZARD:
            hazard_appended = "N"
        continue
    else:
        dq.append((timestamp, category))
        if timestamp == CONTROL_NUM + 3:
            weather_at_3_appended = "Y"
        if category == LOCAL_HAZARD:
            hazard_appended = "Y"

# Step 5: Sum timestamps
total_sum = sum(t for t, _ in dq)

# Step 6: Output
print("LOCAL_HAZARD:", LOCAL_HAZARD)
print("Burst Size:", len(burst))
print("WEATHER at +3 appended:", weather_at_3_appended)
print("Hazard appended:", hazard_appended)
print("Final Deque Length:", len(dq))
print("Sum of timestamps:", total_sum)
print("Final Deque:", dq)
```
